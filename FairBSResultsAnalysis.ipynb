```python
import pandas as pd
import numpy as np
import glob
```

```python
# Set the path to the directory containing the CSV files
path = './results/'

compilation_path = f'{path}compilation/'

# Find all CSV files in the directory
csv_files = glob.glob(path + '/*.csv')

# Create an empty list to store the DataFrames
result_dfs = []

# Loop through the CSV files and append them to the list
for file in csv_files:
    result_df = pd.read_csv(file)
    result_dfs.append(result_df)

# Concatenate the DataFrames into a single DataFrame
df = pd.concat(result_dfs, ignore_index=True)

df = df.drop(columns=['timestamp'])

# Print the combined DataFrame
df
```

```python
df_selected = df[(df['approach_name'].isin(['AEQUITAS', 'FairBS', 'SDG'])) & (df['classifier_name'].isin(['dt', 'rf', 'mlp', 'lr']))]
df_selected
```

```python
df_selected = df_selected.groupby(['approach_name', 'dataset_name', 'sensitive_name', 'classifier_name']).mean().reset_index()

# Round all numeric columns to 2 decimal places
df_selected = df_selected.round(2)

# Specify the columns you want to round to the nearest whole number
columns_to_round = ['tot_inputs', 'disc_inputs'] 
# Round the specified columns to 0 decimal places
df_selected[columns_to_round] = df_selected[columns_to_round].round(0)

df_selected
```

```python
approaches = ['FairBS', 'SDG', 'AEQUITAS']
# approaches =  df_selected['approach_name'].unique()
for approach in approaches:
    df_selected[(df_selected['approach_name']==approach)].to_csv(f'{compilation_path}{approach}_results.csv', index=False)
```

```python
grouped_df = df_selected.groupby(['approach_name', 'dataset_name', 'sensitive_name', 'classifier_name']).mean()

# Pivot the DataFrame so that each approach's metrics become separate columns
pivot_df = grouped_df.unstack(level=0)

# Flattening the multi-level columns
pivot_df.columns = ['_'.join(col).strip() for col in pivot_df.columns.values]

# Resetting index to bring 'dataset_name', 'sensitive_name', and 'classifier_name' back into the DataFrame
pivot_df.reset_index(inplace=True)

pivot_df.to_csv(f'{compilation_path}results.csv', index=False)

# Displaying the final DataFrame
pivot_df
```

```python
classifiers = pivot_df['classifier_name'].unique()
for classifier in classifiers:
    pivot_df[(pivot_df['classifier_name']==classifier)].to_csv(f'{compilation_path}{classifier}_results.csv', index=False)
```
